# Custom Stopwords generated from given words directory

```python
# Import Module
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
import os

# Folder Path
path = r"C:\Users\Mokksh\PycharmProjects\stopwords"

# Change the directory
os.chdir(path)

# Read text File


def read_text_file(file_path):
	with open(file_path, 'r') as f:
		# print(f.read())
		data = f.read()
		data_into_list = data.split("\n")
		stpwrd = nltk.corpus.stopwords.words('english')
		stpwrd.extend(data_into_list)
		#print(data_into_list)
		return stpwrd


	# iterate through all file
for file in os.listdir():
	# Check whether file is in text format or not
	if file.endswith(".txt"):
		file_path = f"{path}\{file}"

		# call read text file function
		Stopwords=read_text_file(file_path)
print(Stopwords)
```

# Transforming the text by avoiding the Stopwords

```python
import re
def transform(text):

    review = re.sub('[^a-zA-Z0-9]', ' ',text)  # except small and capital letters and numeric remove everythong.
    review = review.lower()                    # lower it.
    review = review.split()
    
    review = [word for word in review if not word in Stopwords]   # remove stopwords.
    review = ' '.join(review)
    return review


df['Transform_Text'] = df['Text'].apply(transform)
```

# Syllable Count

```python
def syllable_count(x):
    v = []
    d = {}
    for i in x:
        if i in "aeiou":
            v.append(i)
            d[i] = d.get(i,0)+1
            
    k = []
    for i in d:
        k.append(d[i])

    return np.sum(k)
```

```python
df['syllable count'] = df['Transform_Text'].apply(lambda x: syllable_count(x)) # single liner
```

# Counting the Complex word by checking there syllable count

```python
from collections import  Counter

def complex_word_count(x):
    
    syllable = 'aeiou'
    
    t = x.split()
    
    v = []
    
    for i in t:
        words = i.split()
        c=Counter()
        
        for word in words:
            c.update(set(word))

        n = 0
        for a in c.most_common():
            if a[0] in syllable:
                if a[1] >= 2:
                    n += 1
                
        m = 0
        p = []
        for a in c.most_common():
            if a[0] in syllable:
                p.append(a[0])
        if len(p) >= 2:
            m += 1
        
        if n >= 1 or m >= 1:
            v.append(i)
            
    return len(v) 
```

```python
df['complex_count'] = np.nan
df['complex_count'] = df['Transform_Text'].apply(lambda x: complex_word_count(x))
```

# Analysis of Readability
Analysis of Readability is calculated using the Gunning Fox index formula described below.(mentioned to be used)

Average Sentence Length = the number of words / the number of sentences

Percentage of Complex words = the number of complex words / the number of words

Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)

```python
df['sentence length'] = np.nan
df['Average Sentence Length'] = np.nan
df['Percentage of Complex words'] = np.nan
df['Fog Index'] = np.nan


for i in range(0,len(df)):
    if i not in [35,48]:
        df['sentence length'][i]  =   len(nltk.sent_tokenize(df['Text'][i]))
        df['Average Sentence Length'][i] = df['word_counts'][i]/df['sentence length'][i]
        df['Percentage of Complex words'][i] = df['complex_count'][i]/df['word_counts'][i] 
        df['Fog Index'][i] = 0.4 * (df['Average Sentence Length'][i] + df['Percentage of Complex words'][i])
```

# Making list of positive and negative words using giving txt files

```python
def Dict_text_file(file_path):
	with open(file_path, 'r') as f:
		# print(f.read())
		data = f.read()
		data_into_list = data.split("\n")
		return data_into_list
```

```python
P = r"C:\Users\Mokksh\PycharmProjects\Dictionary_P_N\positive-words.txt"
N = r"C:\Users\Mokksh\PycharmProjects\Dictionary_P_N\negative-words.txt"

Positive = Dict_text_file(P)
Negative = Dict_text_file(N)
```

# Positive Score

```python
def positive_score(x):
    
    s = x.split()
    positive = 0
    
    for i in s:
        if i in Positive:
            positive += 1
    
    #print(positive)
    return positive
```

```python
df['positive_score'] = np.nan

for i in range(len(df)):
    if i not in [35,48]:
            df['positive_score'][i] = positive_score(df['Transform_Text'][i])
```

# Negative Score

```python
def negative_score(x):
    
    s = x.split()
    negative = 0
    
    for i in s:
        if i in Negative:
            negative += 1
    
    #print(positive)
    return negative
```

```python
df['negative_score'] = np.nan

for i in range(len(df)):
    if i not in [35,48]:
            df['negative_score'][i] = negative_score(df['Transform_Text'][i])
```

```python
df.head()
```

# Polarity Score

Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)

```python
df['Polarity Score'] = np.nan

for i in range(len(df)):
    if i not in [35,48]:
        df['Polarity Score'][i] = (df['positive_score'][i]-df['negative_score'][i])/ ((df['positive_score'][i] + df['negative_score'][i]) + 0.000001)
```
